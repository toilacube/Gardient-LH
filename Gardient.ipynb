```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
#X=np.array([80, 64, 30, 57, 80, 28, 100, 44, 55,  78])
#Y=np.array([7.8, 4.3, 1.8, 5, 8.1, 3.5, 10.3, 6, 6.7, 7])
data = pd.read_csv('Nhung_data.csv')
X=data.iloc[:, 0]
Y=data.iloc[:, 1]
plt.scatter(X,Y)
```

```python
m=0
c=0
```

```python
L=0.0001
```

```python
step=1000
```

```python
n=float(len(X))
```

```python
for i in range(step):
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
print (m, c) 
```

```python
a=0.09453
b=0.2267
Y_pred = m*X + c
x=np.linspace(0.,100.)
plt.scatter(X, Y) 
plt.plot(x,a*x+b,'orange',label='Partial derivative')        # Using partial derivative
plt.plot(X,Y_pred, color='green',label='Gardient') # Using Gardient
plt.legend(loc='upper center')
plt.show()
```
